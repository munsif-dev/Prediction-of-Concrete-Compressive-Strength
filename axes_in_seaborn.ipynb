```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
fig, axes = plt.subplots(1, 3, figsize=(12, 6))  # Create 1 row, 2 columns of Axes

# First plot in the left Axes
sns.histplot(data=tips, x="total_bill", ax=axes[1]).set_title("Histogram of Total Bill")


# Second plot in the right Axes
sns.histplot(data=tips, x="tip",kde=True, ax=axes[2]).set_title("Histogram with KDE")
sns.histplot(data=tips,hue=True, x="tip",kde=True, ax=axes[0]).set_title("Histogram with HUE")

plt.tight_layout()
plt.show()
```
