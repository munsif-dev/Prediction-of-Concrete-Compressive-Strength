```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.normal(size=1000)

plt.figure(figsize=(10, 6))

# Basic histogram
sns.histplot(data, bins=10, kde=True, color="skyblue", stat="density", element="bars", fill=True)
plt.title("Histogram with KDE")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import pandas as pd

# Generate sample data
df = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0, 1, 500), np.random.normal(3, 1, 500)]),
    'category': ['A'] * 500 + ['B'] * 500
})

plt.figure(figsize=(10, 6))

# Histogram with hue
sns.histplot(data=df, x="value", hue="category", multiple="dodge", bins=30, alpha=0.7)
plt.title("Histogram with Categorical Hue")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```
