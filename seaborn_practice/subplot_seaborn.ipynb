```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 6))

# First subplot (1st row, 2nd column, 1st plot)
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Sine Wave")

# Second subplot (1st row, 2nd column, 2nd plot)
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine Wave")

# Third subplot (2nd row, 2nd column, 3rd plot)
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title("Tangent Wave")

# Fourth subplot (2nd row, 2nd column, 4th plot)
plt.subplot(2, 2, 4)
plt.plot(x, -np.sin(x))
plt.title("Negative Sine Wave")

plt.tight_layout()  # Adjusts layout to prevent overlapping
plt.show()
```
